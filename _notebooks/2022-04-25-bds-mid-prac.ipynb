### `2`. 

````python
x = tnp.array([11.0,12.0,13.0,14.0,15.0])
y = tnp.array([17.7,18.5,21.2,23.6,24.2])
````

(a) ëª¨í˜• $y_i=\beta_0+\beta_1 x_i$ ì— í•´ë‹¹í•˜ëŠ” ë„¤íŠ¸ì›Œí¬ë¥¼ kerasë¥¼ ì´ìš©í•˜ì—¬ ì„¤ê³„í•˜ê³  ì†ì‹¤í•¨ìˆ˜ë¥¼ ì •ì˜í•˜ë¼. $(\beta_0,\beta_1)=(3,3)$ ì¼ ê²½ìš°ì˜ lossë¥¼ ê³„ì‚°í•˜ë¼.
- ì†ì‹¤í•¨ìˆ˜ëŠ” MSELossë¥¼ í™œìš©í•œë‹¤.

````python
X = tf.concat([tf.ones(5,dtype=tf.float64).reshape(5,1),x.reshape(5,1)],axis=1)
y = y.reshape(5,1) 
````

````python
X
````

````python
X.T
````

````python
beta = tf.linalg.inv(X.T @ X) @ X.T @ y
beta
````

````python
loss = -2*X.T @ y + 2*X.T@X@beta 
loss
````

````python
beta = tf.Variable(tnp.array([3.0,3.0]).reshape(2,1))
beta
````

````python
loss = -2*X.T @ y + 2*X.T@X@beta 
loss
````

(b) $(\beta_0,\beta_1)=(3,3)$ ì—ì„œ ì†ì‹¤í•¨ìˆ˜ì˜ ë¯¸ë¶„ê³„ìˆ˜ë¥¼ ê³„ì‚°í•˜ë¼.

````python
with tf.GradientTape() as tape:
    tape.watch(beta)
    yhat = X @ beta
    loss = (y-yhat).T@(y - yhat)
````

````python
tape.gradient(loss, beta)
````

````python
(y-X @ beta).T@(y - X @ beta)
````

(c) ê²½ì‚¬í•˜ê°•ë²•ì„ í†µí•˜ì—¬ $(\beta_0,\beta_1)=(3,3)$ ì˜ ê°’ì„ 1íšŒ updateí•˜ë¼. ì—¬ê¸°ì—ì„œ í•™ìŠµë¥ ì€ 0.01ë¡œ ì„¤ì •í•œë‹¤.

````python
opt = tf.keras.optimizers.SGD(0.01)
````

````python
with tf.GradientTape(persistent=False) as tape:
    tape.watch(beta)
    yhat = X @ beta
    loss = (y-yhat).T@(y-yhat)/5
slope = tape.gradient(loss,beta)
opt.apply_gradients([(slope,beta)])
````

````python
tf.random.set_seed(43052)
epsilon=tf.random.normal([10])
X = tnp.array([1]*10 +[20.1, 22.2, 22.7, 23.3, 24.4, 25.1, 26.2, 27.3, 28.4, 30.4]).reshape(2,10).T
beta = tnp.array([9.0,2.0]).reshape(2,1)
beta_true= tnp.array([10.2,2.2]).reshape(2,1)
y= X@beta_true+epsilon.reshape(10,1) 
````

````python
with tf.GradientTape(persistent=True) as tape: 
    tape.watch(beta)
    yhat= X@beta
    loss= (y-yhat).T @(y-yhat) 
````

ë³€ìˆ˜ê°€ Variableë¡œ ìƒì„±ë˜ë©´ ìë™ì ìœ¼ë¡œ ê´€ì°°ë˜ì§€ë§Œ constantë¡œ ìƒì„±ë˜ë©´ ê´€ì°°ë˜ì§€ ì•Šìœ¼ë‹ˆ ë³´ê³  ì‹¶ë‹¤ë©´ watch ì¶”ê°€í•˜ê±°ë‚˜ variableë¡œ ìƒì„±í•˜ë©´ ëœë‹¤.

````python
tape.gradient(loss,beta) # tf ê¸°ìš¸ê¸° ë¯¸ë¶„í•œ ê³„ì‚°ê°’
````

````python
beta_optimal = tf.linalg.inv(X.T @ X) @ X.T  @y  # ì´ë³´ë‹¤ lossë¥¼ ì‘ê²Œ ë§Œë“œëŠ” betaëŠ” ì—†ë‹¤.
````

````python
with tf.GradientTape(persistent=True) as tape: 
    tape.watch(beta_optimal)
    yhat= X@beta_optimal
    loss= (y-yhat).T @(y-yhat) 
tape.gradient(loss,beta_optimal) # ë² íƒ€ ìµœì ê°’
````
